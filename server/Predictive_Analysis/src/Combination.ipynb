```python
plt_data  = pd.read_csv("nyc_census_tracts.csv")
city_data = pd.read_csv("census_block_loc.csv")
competition_data = pd.read_csv("Retail_Food_Stores.csv")
```

```python
cols_to_keep = ['County','Borough','TotalPop','Income','IncomeErr', 'IncomePerCap', 'IncomePerCapErr','Poverty','Employed']

pop_data = plt_data[cols_to_keep].copy()
```

```python
selected_cities = ['Bronx', 'Kings', 'New York', 'Queens', 'Richmond']
filter_county = city_data[city_data['County'].isin(selected_cities)]
filter_county.info()
filter_county['County'].unique()
```

```python
filter_county = filter_county.drop(columns=['BlockCode','State'])
filter_county.head()
```

County Data according to the latitude and longitude

```python
richmond_coord = filter_county[filter_county['County'] == 'Richmond'].copy()
queens_coord = filter_county[filter_county['County'] == 'Queens'].copy()
kings_coord = filter_county[filter_county['County'] == 'Kings'].copy()
ny_coord = filter_county[filter_county['County'] == 'New York'].copy()
bronx_coord = filter_county[filter_county['County'] == 'Bronx'].copy()
```

County Data for Poplulation

```python
richmond_pop = pop_data[pop_data['County'] == 'Richmond'].dropna() # 107
queens_pop = pop_data[pop_data['County'] == 'Queens'].dropna() # 642
kings_pop = pop_data[pop_data['County'] == 'Kings'].dropna() # 748
ny_pop = pop_data[pop_data['County'] == 'New York'].dropna() # 275
bronx_pop = pop_data[pop_data['County'] == 'Bronx'].dropna() # 329
```

```python
queens_pop.info()
```

```python
seed = 42
bronx_coord_samp = bronx_coord.sample(n=329, random_state=seed)
bronx_coord_samp.head()
bronx_coord_samp.info()
```

```python
combined_bronx = pd.merge(bronx_pop,bronx_coord_samp,on='County')
combined_bronx.head()
```

```python
combined_bronx.to_csv("Bronx_Combined.csv")
```

New York

```python
seed = 42
ny_coord_samp = ny_coord.sample(n=275, random_state=seed)
ny_coord_samp.head()
ny_coord_samp.info()
```

```python
combined_ny = pd.merge(ny_pop,ny_coord_samp,on='County')
combined_ny.head()
```

```python
combined_ny.to_csv("NY_combined.csv")
```

Kings

```python
seed = 42
kings_coord_samp = kings_coord.sample(n=748, random_state=seed)
kings_coord_samp.head()
kings_coord_samp.info()
```

```python
combined_kings = pd.merge(kings_pop,kings_coord_samp,on='County')
combined_kings.head()
```

```python
combined_kings.to_csv("Kings_Combined.csv")
```

Queens

```python
seed = 42
queens_coord_samp = queens_coord.sample(n=642, random_state=seed)
queens_coord_samp.head()
# queens_coord_samp.info()
```

```python
combined_queens = pd.merge(queens_pop,queens_coord_samp,on='County')
combined_queens.head()
combined_queens.describe()
```

```python
combined_queens.to_csv("Queens_Combined.csv")
```

RichMond

```python
seed = 42
richmond_coord_samp = richmond_coord.sample(n=107, random_state=seed)
richmond_coord_samp.head()
richmond_coord_samp.info()
```

```python
combined_richmond = pd.merge(richmond_pop,richmond_coord_samp,on='County')
```

```python
combined_richmond.head()
combined_richmond.describe()
```

```python
combined_richmond.to_csv("RichMond_Combined.csv")
```

```python
combined_df = pd.concat([combined_richmond, combined_kings, combined_bronx,combined_queens,combined_ny],axis=0)
```
