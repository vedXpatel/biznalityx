```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/food_delivery_data.csv')
df.head()
```

**Maker Map for all the orders**

```python
# Create a map
m_2 = folium.Map(location=[19.0047	,72.8415	], tiles='cartodbpositron', zoom_start=13)

for idx, row in df.iterrows():
    Marker([row['DeliveryLocation(Latitude)'], row['DeliveryLocation(Longitude)']]).add_to(m_2)

m_2
```

```python
# Create the map
import math
m_3 = folium.Map(location=[19.0047	,72.8415], tiles='cartodbpositron', zoom_start=13)

# Add points to the map
mc = MarkerCluster()
for idx, row in df.iterrows():
    if not math.isnan(row['DeliveryLocation(Longitude)']) and not math.isnan(row['DeliveryLocation(Latitude)']):
        mc.add_child(Marker([row['DeliveryLocation(Latitude)'], row['DeliveryLocation(Longitude)']]))
m_3.add_child(mc)

# Display the map
m_3
```

**HeatMap for all Order**

```python
# Create a base map
m_5 = folium.Map(location=[19.0047,72.8415], tiles='cartodbpositron', zoom_start=12)

# Add a heatmap to the base map
HeatMap(data=df[['DeliveryLocation(Latitude)', 'DeliveryLocation(Longitude)']], radius=10).add_to(m_5)

# Display the map
m_5
```

```python
# Create map with release incidents and monitoring stations
m = folium.Map(location=[19.0047,72.8415], zoom_start=11)
HeatMap(data=df[['DeliveryLocation(Latitude)', 'DeliveryLocation(Longitude)']], radius=15).add_to(m)
for idx, row in df.iterrows():
    Marker([row['DeliveryLocation(Latitude)'], row['DeliveryLocation(Longitude)']]).add_to(m)

m
```
